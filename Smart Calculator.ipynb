```python
from tkinter import *

root= Tk()
root.title("Claculator")
root.geometry('380x470+200+100')
root.resizable(False,False)
# root.configure(bg='#133337')

#--------Functions-----------

def enterNumber(x):
    if entry_box.get() == '0':
        if x == '.':
            entry_box.insert(1,'.')
            
        
        else:
            entry_box.delete(0,'end')
            entry_box.insert(0,str(x))
        
    else:
        lenght = len(entry_box.get())
        entry_box.insert(lenght, str(x))
        
        
def enter_oprators(x):
    if entry_box.get() != '0':
        lenght = len(entry_box.get())
        all_text = entry_box.get()
        last_char = all_text[-1:]
        
        if last_char in ['+', '-', '/'] or all_text[-2:] == '**':
            pass
        else:
            entry_box.insert(lenght, btn_oprators[x]['text'])
        
def funclear():
    entry_box.delete(0,END)
    entry_box.insert(0,'0')
    
    
result = 0
history = []
def func_output():
    content = entry_box.get()
    result = eval(content)
    entry_box.delete(0,END)
    entry_box.insert(0, result)
    history.append(content)
    history.reverse()
    status.configure(text="Histroy: " + '|'.join(history[0:4]), font='verdana 10 bold')
    

def func_Delete():
    lenght= len(entry_box.get())
    entry_box.delete(lenght-1,'end')
    if lenght == 1:
        entry_box.insert(0,'0')

#-------entrybox-----------

entry_box = Entry(font='verdana 14 bold', width=18, bd=5, justify= RIGHT, bg='#e6e6fa')
entry_box.insert(0,'0')
entry_box.place(x =30, y =10)

#----------button numbers----------

btn_number =[]
for i in range(10):
    btn_number.append(Button(width=4, text= str(i), bd=4,command=lambda x =i:enterNumber(x)))
    
    
btn_text = 1    
for i in range(0,3):
    for j in range(0,3):
        btn_number[btn_text].place(x=30+j*90, y =70+i*70)
        btn_text+=1

        
#-------------Zero Button------------
btn_zero = Button(width = 31, text='0', bd=4, command= lambda x =0:enterNumber(x))
btn_zero.place(x=25, y=280)

#-------------DOT button------------
btn_dot = Button(width = 4, text='.', font='times 15 bold', bd=4, command= lambda x ='.':enterNumber(x))
btn_dot.place(x=110, y=340)

#-------------Status bar--------------
status = Label(root, text='Histroy:', relief=SUNKEN, height=2, anchor=W, font='verdana 11 bold')
status.pack(side=BOTTOM,fill=X)




#------------button oprators----------

btn_oprators = []

for i in range(4):
    
    btn_oprators.append(Button(width =4, text = 'Times 14 bold', bd = 5, command = lambda x =i:enter_oprators(x)))
                    
btn_oprators[0]['text'] = '+'
btn_oprators[1]['text'] = '-'
btn_oprators[2]['text'] = '*'
btn_oprators[3]['text'] = '/'

for i in range(4):
    btn_oprators[i].place(x =290, y =70+i*70)
    
    
#-------------CLEAR button----------
btn_clear = Button(width =4, text='C', font='times 15 bold',bd=5, command= funclear)
btn_clear.place(x=25, y=340)


#---------------Equal button-----------
btn_equal = Button(width = 4, text='=', font='times 15 bold', bd=4, command= func_output)
btn_equal.place(x=200, y=340)


#----------------Delete Button -----------------
btn_del = Button(width = 4, text='Del', font='times 15 bold', command=func_Delete)
btn_del.place(x=290, y=340)



root.mainloop() 
```
