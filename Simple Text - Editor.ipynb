```python
from tkinter import *
from tkinter import filedialog


class text_editor():
    
   #define the open_file function for opening the file in TextPad
   

    def open_file(self):
        open_return = filedialog.askopenfile(initialdir="/", title= "Select file to open", filetype =(("text files", "*.txt"), ("all files", "*.*")))
        
        #if nothing is opean previou text a it is
        if(open_return!=None):     
            self.text_area.delete(1.0, END) 
        #it's delete the previous open file and look for new one.
        
        # this is define for open file to TextPad by using insert function 
        for line in open_return:
            self.text_area.insert(END, line)
        open_return.close()
    
    def  save_as_file(self):
        f = filedialog.asksaveasfile(mode="w", defaultextension=".txt")
        if f is None:
            return
        text2save = self.text_area.get(1.0, END)
        f.write(text2save)
        f.close()
    
    
    def __init__(self, master):      #self is used to represent the variable defined
        self.master = master        #here master work as root window
        master.title("TextPad")
        self.text_area = Text()
        self.text_area.pack(fill=BOTH, expand=1)
        
        #creating menu bar in TextPad using self function        
        self.main_menu = Menu()
        self.master.config(menu = self.main_menu)                 #for presenting menu bar on window using config
        
        
        #creating file menu
        self.file_menu = Menu(self.main_menu)
        self.main_menu.add_cascade(label="File", menu=self.file_menu)  # add_cascade is used for add menu bar in main menu
        self.file_menu.add_command(label="Open", command=self.open_file)
        self.file_menu.add_command(label="Save as", command=self.save_as_file)
        
        #creating edit menu
        self.edit_menu = Menu(self.main_menu)
        self.main_menu.add_cascade(label="Edit", menu=self.edit_menu)
        
        
        #creating view menu
        self.view_menu = Menu(self.main_menu)
        self.main_menu.add_cascade(label="View", menu=self.view_menu)
        
        
        #creating Help menu
        self.help_menu = Menu(self.main_menu)
        self.main_menu.add_cascade(label="Help", menu=self.help_menu)
    





root = Tk()
editor = text_editor(root)
root.mainloop()
```
